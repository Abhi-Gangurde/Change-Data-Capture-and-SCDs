# Initializaton

```python
from pyspark.sql.functions import lit
```

### Existing Dimension

```python
dim_df = spark.createDataFrame(
    [
        [1, "Ravi", "Mumbai"],
        [2, "Anil", "Delhi"]
    ],
    ["customer_id", "customer_name", "city"]
)
```

```python
dim_df.show()
```

### New incomming dat

```python
update_df = spark.createDataFrame([
                                  (1,"Ravi","Pune")
                                  ],
                                  ["customer_id","customer_name","city"]

                                  )
```

```python
update_df.show()
```

```python
# Type 1: Overwrite 

#remove records which needs to be updated 

filder_dim_df = dim_df.join(update_df, on="customer_id", how="left_anti")

#union with dim_df for the new records
final_df = filder_dim_df.union(update_df)
final_df.show()
```

```python
#pyspark when matched and unmatched
from pyspark.sql.functions import col

dim_df.alias("d").join(update_df.alias("u"), on="customer_id", how="outer").select(
    col("d.*"),
    col("u.*").when(col("u.customer_id").isNotNull(), col("u.*")).otherwise(lit(None)).alias("updated")
).show()
```
